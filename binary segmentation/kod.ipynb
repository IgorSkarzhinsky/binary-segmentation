```
import torch
from torch.utils.data import Dataset
import cv2
import numpy as np

class SegmentationDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Загружаем изображение
        img = cv2.imread(self.image_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (256, 256))
        img = img.astype(np.float32) / 255.0
        img = torch.from_numpy(img).permute(2, 0, 1)

        # Загружаем маску
        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (256, 256), interpolation=cv2.INTER_NEAREST)
        mask = torch.from_numpy(mask).long()

        return img, mask
```

```
import torch
from torch.utils.data import Dataset
import cv2
import numpy as np
import glob

class SegmentationDataset(Dataset):
    def __init__(self, image_paths, mask_paths):
        self.image_paths = image_paths
        self.mask_paths = mask_paths

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        # Загрузка и нормализация изображения
        img = cv2.imread(self.image_paths[idx])
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (256, 256))
        img = img.astype(np.float32) / 255.0
        img = torch.from_numpy(img).permute(2, 0, 1)

        # Загрузка и преобразование маски
        mask = cv2.imread(self.mask_paths[idx], cv2.IMREAD_GRAYSCALE)
        mask = cv2.resize(mask, (256, 256), interpolation=cv2.INTER_NEAREST)
        # ПРЕОБРАЗОВАНИЕ: 29 -> 0, 149 -> 1
        mask = np.where(mask == 29, 0, 1)  # Здесь важен порядок
        mask = torch.from_numpy(mask).long()

        return img, mask
```

```
import torch.optim as optim
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
import segmentation_models_pytorch as smp

# Пути к данным
base_path = "/content/NVIDIA-Aerial-Drone-Dataset/FPV/SFWA/720p"
images = sorted(glob.glob(f"{base_path}/images/*.png"))
masks = sorted(glob.glob(f"{base_path}/labels/*.png"))

# Разделение на train/val
train_imgs, val_imgs, train_masks, val_masks = train_test_split(
    images, masks, test_size=0.2, random_state=42
)

# Создание датасетов и загрузчиков
train_dataset = SegmentationDataset(train_imgs, train_masks)
val_dataset = SegmentationDataset(val_imgs, val_masks)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False)

# Модель (2 класса: небо и земля)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = smp.Unet(
    encoder_name="resnet34",
    encoder_weights="imagenet",
    in_channels=3,
    classes=2  # <-- ИЗМЕНЕНО НА 2
).to(device)

# Обучение
criterion = torch.nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)

num_epochs = 5  # Для быстрого теста
for epoch in range(num_epochs):
    model.train()
    train_loss = 0
    for imgs, masks in train_loader:
        imgs, masks = imgs.to(device), masks.to(device)
        optimizer.zero_grad()
        outputs = model(imgs)
        loss = criterion(outputs, masks)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    print(f"Epoch {epoch+1}, Loss: {train_loss/len(train_loader):.4f}")
```

```
import matplotlib.pyplot as plt
import numpy as np

def show_prediction(model, img, mask):
    model.eval()
    with torch.no_grad():
        img_tensor = img.unsqueeze(0).to(device)
        pred = model(img_tensor)
        pred_mask = torch.argmax(pred, dim=1).squeeze().cpu().numpy()

    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes[0].imshow(img.permute(1, 2, 0))
    axes[0].set_title("Изображение с дрона")
    axes[1].imshow(mask, cmap='gray')
    axes[1].set_title("Истинная маска (0=земля, 1=небо)")
    axes[2].imshow(pred_mask, cmap='gray')
    axes[2].set_title("Предсказание модели")
    plt.show()

# Покажем на первых 3 примерах из валидации
for i in range(3):
    img, mask = val_dataset[i]
    show_prediction(model, img, mask)
```
